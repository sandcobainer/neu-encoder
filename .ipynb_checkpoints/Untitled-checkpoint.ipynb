```python
import os
import fnmatch
import numpy as np
import matplotlib.pyplot as plt
from magenta.models.nsynth import utils
from magenta.models.nsynth.wavenet import fastgen
from IPython.display import Audio
%matplotlib inline
%config InlineBackend.figure_format = 'jpg'
```

load_encoding function takes in the parameters and returns the loaded audio and corresponding encodings

```python
def load_encoding(fname, sample_length=None, sr=20000, ckpt='./wavenet-ckpt/model.ckpt-200000'):
    audio = utils.load_audio(fname, sample_length=sample_length, sr=sr)
    encoding = fastgen.encode(audio, ckpt, sample_length)
    return audio, encoding
```

```python
# from https://www.freesound.org/people/maurolupo/sounds/213259/
script_dir = os.path.dirname(os.path.realpath('__file__'))
raw_samples = './audio/'
rel_path = os.path.join(script_dir, raw_samples)
gen_samples = './gen_audio/'
sample_length = 5000

for root, dirs, files in os.walk(raw_samples):  
    for fname in files:
        if fnmatch.fnmatch(fname, '*.wav'):
            print(fname);
            audio, encoding = load_encoding(rel_path + fname, sample_length)
            np.save(fname + '.npy', encoding)
            %time fastgen.synthesize(encoding,save_paths=['./gen_audio/' + fname],checkpoint_path='./wavenet-ckpt/model.ckpt-200000',samples_per_save=sample_length)

# fastgen.synthesize(
#     encoding,
#     save_paths=['./audio' + '/gen_/' + fname],
#     samples_per_save=sample_length)
# synthesis = utils.load_audio( '/gen_audio' + fname,
#                              sample_length=sample_length,
#                              sr=sr)
```
